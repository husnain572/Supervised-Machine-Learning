## 3. Relational Plot

- `relplot()` in Seaborn is a figure-level function used to visualize relationships between two numerical variables.
- It creates either a **scatter plot (default)** or a **line plot** using `kind="scatter"` or `kind="line"`.
- It is mainly used for **numerical vs numerical** analysis and supports grouping with categorical variables using `hue`, `style`, `col`, or `row`.
- It is useful in EDA to understand trends, patterns,correlation and time series and comparisons across categories.


```python
sns.relplot(
    data=tips,
    y='tip',
    x='total_bill')
plt.title("Relational Plot: Total Bill vs Tip")
plt.show()
```

```python
sns.relplot(
    data=tips,
    y='tip',
    x='total_bill',
    hue='sex',  # categorical grouping
    size='size',
    style='sex'
)
plt.title("Relational Plot: Total Bill vs Tip")
plt.show()
```

```python
# Line relational plot (time series)
sns.relplot(
    data=flights,
    x='year',
    y='passengers',
    kind='line'
    )
plt.title("Line Plot: Passengers Over Years")
plt.show()
```

## 4. SCATTER PLOT

- `scatterplot()` in Seaborn is an axes-level function used to visualize the relationship between two numerical variables.
- It displays data as individual points to help identify patterns, trends, correlations, or outliers.
- It is mainly used for **numerical vs numerical** analysis.
- We can also use `hue`, `style`, and `size` to represent additional categorical or numerical variables.

```python
sns.scatterplot(
    data=tips,
    y='tip',
    x='total_bill',
    hue='sex',

)
plt.title("Scatter Plot between tips and total_bill")
plt.show()
```

## 5. Line Plot

- `lineplot()` in Seaborn is an axes-level function used to show the relationship between two variables using lines.
- It is mainly used for **time series or continuous data** to visualize trends over an interval.
- Typically, the x-axis represents a continuous variable (like time) and the y-axis represents a numerical variable.
- It also supports `hue`, `style`, and `size` to compare trends across different categories.


```python
sns.lineplot(
    data=flights,
    x="year",
    y="passengers",
    ci='sd'             # confidence interval
)
plt.title("Line Plot with Confidence Interval")
plt.show()
```

## 6. CATEGORICAL PLOTS (Categorical vs Numerical)
### 6.1. Bar Plot (Mean estimator)

- `barplot()` in Seaborn is an axes-level function used to compare **numerical values across categories**.
- It usually shows the **mean (average)** of a numerical variable for each category as bars.
- It is mainly used for **categorical vs numerical** analysis.
- By default, it also shows a **confidence interval** as a small error bar on top of each bar.

- Use it when you want to compare averages between different groups.


```python
sns.barplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    estimator=np.mean
)
plt.title("Bar Plot: Mean Total Bill per Day")
plt.show()
```

### 6.2. Count Plot
- `countplot()` in Seaborn is an axes-level function used to **show the number of occurrences of each category**.
- It is mainly used for **categorical data** and does **not require a numerical variable**, as it counts the instances automatically.
- You can also use `hue` to see counts across sub-categories.
- Use it when you want to **visualize frequency distribution** of categories.


```python
sns.countplot(
    data=tips,
    x='day',
    hue='sex'
)
plt.title("Count Plot: Frequency of Visits")
plt.show()
```

### 6.3. Box Plot

`boxplot()` in Seaborn is an axes-level function used to **visualize the distribution of numerical data** across categories.
It shows:

* **Median** (central line)
* **Quartiles** (box edges)
* **Whiskers** (range of most data)
* **Outliers** (points outside whiskers)

It is mainly used for **categorical vs numerical** data to understand spread, skewness, and detect outliers.

You can also use `hue` to compare distributions across sub-categories.


```python
sns.boxplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex'
)
plt.title("Box Plot: Distribution per Day")
plt.show()
```

### 6.4. Violin Plot

`violinplot()` in Seaborn is an axes-level function used to **visualize the distribution of numerical data across categories**, similar to `boxplot()`.

* It shows a **smoothed density** of the data (like a rotated kernel density plot) instead of just quartiles.
* Median and interquartile range can also be displayed inside the violin.
* Mainly used for **categorical vs numerical** data to understand **distribution shape, spread, and outliers**.
* Supports `hue` to compare distributions across sub-categories.

In short: **violinplot = boxplot + density information**.


```python
sns.violinplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    split=True

)
```

### 6.5. Strip Plot
`stripplot()` in Seaborn is an axes-level function used to **show individual data points for a categorical variable**.

* X-axis → categorical variable
* Y-axis → numerical variable
* Each point represents one observation
* Useful for seeing **all individual values** and **distribution of data**

Extra features:

* `hue` → color points by category
* `jitter=True` → adds small random noise to avoid overlapping points

Short: **stripplot = scatter plot for categorical data, showing all points individually**.


```python
sns.stripplot(
    data=tips,
    x="day",
    y="total_bill",
    jitter=True
)
plt.title("Strip Plot: Individual Data Points")
plt.show()
```

### 6.6. Swarm Plot

`swarmplot()` in Seaborn is an axes-level function similar to `stripplot()`, but it **automatically arranges points to avoid overlap**.

* X-axis → categorical variable
* Y-axis → numerical variable
* Each point represents one observation
* Unlike `stripplot()`, you **don’t need `jitter`**, because Seaborn spreads points intelligently

Extra features:

* `hue` → color points by category
* Good for **small to medium-sized datasets** where you want to **see all points clearly**

Short: **swarmplot = non-overlapping stripplot showing every individual data point clearly**.


```python
sns.swarmplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)
plt.title("Swarm Plot")
plt.show()
```

## 7 Distribution Plot (Single and Two Variables)
###  7.1. Histogram + KDE

- A **Histogram** in Matplotlib is used to visualize the **distribution of numerical data** by dividing it into bins and showing frequency.
- KDE (Kernel Density Estimation) adds a **smooth curve** to represent the data distribution more continuously.
- Histogram shows data using bars, while KDE shows a smooth density line.
- Use them together to clearly understand the **shape, spread, and skewness** of numerical data.


```python
sns.histplot(
    data=tips,
    x='total_bill',
    bins=10,
    kde=True
)
plt.title("Histogram with KDE")
plt.show()
```

### 7.2. KDE Plot (Density Only)

- A **KDE Plot** in Matplotlib shows the **smooth probability density distribution** of numerical data without histogram bars.
- It represents how data is spread using a continuous curve.
- Higher peaks indicate more concentrated values.
- Use it to understand the **shape, skewness, and overall distribution** of numerical data clearly.


```python
sns.kdeplot(
    data=tips,
    x='total_bill',
    fill=True
)
plt.title("KDE Plot")
plt.show()
```

### 7.3. Pair Plot

- `pairplot()` in Seaborn is a figure-level function used to visualize **pairwise relationships between multiple numerical variables** in a dataset.
- It creates scatter plots for every combination of numerical columns and histograms (or KDE) on the diagonal.
- It is mainly used for **multivariate analysis** and understanding correlations.
- You can use `hue` to compare relationships across categories.


```python
sns.pairplot(
   tips,
   hue='sex'

)
```

## 8. Matrix Plot
### 8.1. Heatmap

- `heatmap()` in Seaborn is used to visualize data in a **matrix format using colors**.
- Color intensity represents the magnitude of values (higher value = darker/stronger color).
- It is commonly used to display a **correlation matrix** between numerical variables.
- Use it to quickly identify **patterns, relationships, and strong correlations** in data.


```python
sns.heatmap(tips.corr(numeric_only=True), annot=True, cmap='viridis')
plt.title("Correlation Heatmap")
plt.show()
```

## 9. Regression Plot
### 9.1. Regression line with confidence interval

- `regplot()` in Seaborn is used to visualize the **relationship between two numerical variables along with a regression line**.
- It shows a scatter plot and automatically fits a **linear regression line** to the data.
- It is mainly used for **numerical vs numerical** analysis to understand trend and correlation.
- By default, it also shows a **confidence interval** around the regression line.


```python
sns.regplot(
    data=tips,
    x='total_bill',
    y='tip'
)
plt.title("Regression line with confidence interval")
plt.show()
```

### 9.2 lmplot (Regression + Faceting)

- `lmplot()` in Seaborn is a figure-level function used to create a **scatter plot with a regression line**, similar to `regplot()`.
- The key difference is that it supports **faceting using `col` and `row`**, allowing multiple regression plots by category.
- It is mainly used for **numerical vs numerical** analysis with grouped comparisons.
- By default, it also displays a **confidence interval** around the regression line.


```python
# lmplot with faceting
sns.lmplot(
    data=tips,
    x="total_bill",
    y="tip",
    col="sex",     # creates separate plots for Male and Female
    height=4,
    aspect=2
    )

plt.show()
```

### 9.3. Facet Grid

- `FacetGrid` in Seaborn is a figure-level object used to create **multiple subplots based on categorical variables**.
- It allows you to split data using `row`, `col`, or both, and then map a plot (like scatter, hist, etc.) to each subset.
- It is mainly used for **comparing distributions or relationships across categories**.
- `lmplot()` and `relplot()` internally use `FacetGrid`.



```python
g = sns.FacetGrid(tips, col="time", row="sex")   # Lunch & Dinner separate plots
g.map(sns.scatterplot, "total_bill", "tip")

plt.show()
```
