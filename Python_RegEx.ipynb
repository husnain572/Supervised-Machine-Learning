```python
import re
```

```python
a= 'charlie chachaplin coa and the chocolate factory'
b='husnainmalikawan@gmail.com'
c='hello'
d='xyz,yz,xyzz,xyyz,xxzzy,zyz,xxy'
```

```python
match=re.findall(r'a',b)
print(match)
```

```python
match=re.search(r'l',b)
print(match)
```

```python
match=re.search(r'^husnain',b)
print(match)
```

```python
match=re.search(r'^ali',b)
print(match)
```

```python
match=re.search(r'husnain$',b)
print(match)
```

```python
match=re.search(r'com$',b)
print(match)
```

```python
match=re.findall(r'c.a',a)
print(match)
```

```python
match=re.findall(r'cha|fac',a)
print(match)
```

```python
match=re.search(r'cha|fac',a)
print(match)
```

```python
match=re.search(r'ali|fac',a)
print(match)
```

```python
match=re.findall(r'ch?o',a)
print(match)
```

```python
match=re.findall(r'ch*o',a)
print(match)
```
