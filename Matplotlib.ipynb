# Matplotlib Master Notebook


```python

import numpy as np
import matplotlib.pyplot as plt
```

# 1. BAR PLOT (Categorical vs Numerical)
## Use Case:
- Compare values across categories
- x-axis: Categorical
- y-axis: Numerical

```python
languages = ["Python", "C++", "C", "Java"]
popularity = [85, 70, 50, 75]

plt.figure(figsize=(6,4))
plt.bar(
    languages,                # categorical x-axis
    popularity,              # numerical y-axis
    width=0.4,
    color=['y','b','r','m'],
    edgecolor='black',
    linewidth=2,
    alpha=0.7
)
plt.xlabel('Programming Language')
plt.ylabel('Popularity (%)')
plt.title('Single Bar Plot: Language Popularity')
plt.show()
```

#### GROUPED BAR PLOT (Compare two numerical series per category)

```python
popularity_2023 = [85, 70, 60, 82]
popularity_2024 = [20, 30, 40, 50]

x_pos = np.arange(len(languages))
bar_width = 0.35

plt.figure(figsize=(7,4))
plt.bar(x_pos, popularity_2023, bar_width, label='2023', color='r')
plt.bar(x_pos + bar_width, popularity_2024, bar_width, label='2024', color='y')

plt.xticks(x_pos + bar_width/2, languages, rotation=20)
plt.xlabel('Programming Language (Categorical)')
plt.ylabel('Popularity (%) (Numerical)')
plt.title('Grouped Bar Plot: Year-wise Comparison')
plt.legend()
plt.show()
```

#### HORIZONTAL BAR PLOT (Use when category names are long)

```python
plt.figure(figsize=(6,4))
plt.barh(languages, popularity, color='y', edgecolor='black', alpha=0.7)
plt.xlabel('Popularity (%) (Numerical)')
plt.ylabel('Programming Language (Categorical)')
plt.title('Horizontal Bar Plot')
plt.show()
```

# 2. SCATTER PLOT (Numerical vs Numerical)

## Use Case:
- Show relationship between two numerical variables
- Detect correlation, clusters, outliers
- x-axis: Numerical
- y-axis: Numerical



```python

days = [1,2,3,4,5,6,7]
visits = [2,3,1,5,1,7,6]

plt.figure(figsize=(6,4))
plt.scatter(days, visits, color='r', s=80)
plt.xlabel('Day (Numerical)')
plt.ylabel('Number of Visits (Numerical)')
plt.title('Basic Scatter Plot')
plt.show()
```

#### MULTI-SERIES + SIZE + COLOR MAPPED SCATTER


```python
visits_2 = [1,4,1,5,2,5,7]
point_sizes = [10,30,50,20,50,25,70]
color_values = [10,49,30,29,56,20,30]

plt.figure(figsize=(6,4))
sc = plt.scatter(
    days, visits,
    s=point_sizes,
    c=color_values,          # mapped to colormap
    alpha=0.7,
    marker='*',
    edgecolor='g',
    linewidth=1,
    cmap='viridis'
)
plt.scatter(days, visits_2, s=point_sizes, color='r', alpha=0.6, label='Second Series')
plt.colorbar(sc, label='Color Intensity')
plt.xlabel('Day (Numerical)')
plt.ylabel('Visits (Numerical)')
plt.title('Advanced Scatter Plot')
plt.legend()
plt.show()
```

# 3. HISTOGRAM (Distribution of ONE Numerical Variable)
## Use Case:
- Show frequency distribution
- x-axis: Numerical bins
- y-axis: Frequency (counts)

```python
scores = [14,18,51,21,22,22,37,48,51,43,53,19,23,29,35,15,34,28,26,35,
          22,40,8,25,42,59,40,53,36,19,27,51,40,26,20,59,36,33,46,33,
          40,24,22,55,15,18,57,51,19,17]

bins = [10,20,30,40,50,60]

plt.figure(figsize=(6,4))
plt.hist(scores, bins=bins, color='g', edgecolor='r')
plt.xlabel('Score Range (Numerical)')
plt.ylabel('Frequency')
plt.title('Histogram with Custom Bins')
plt.show()
```

#### AUTO BIN + RANGE CONTROL

```python
plt.figure(figsize=(6,4))
plt.hist(scores, bins='auto', range=(0,80), edgecolor='r')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Histogram with Auto Bins')
plt.show()
```

#### HORIZONTAL HISTOGRAM

```python
plt.figure(figsize=(6,4))
plt.hist(scores, bins='auto', range=(0,80), rwidth=0.7,
         orientation='horizontal', edgecolor='r')
plt.xlabel('Frequency')
plt.ylabel('Score')
plt.title('Horizontal Histogram')
plt.show()
```

# 4. PIE PLOT (Categorical Proportions)

## Use Case:
- Show percentage contribution
- Best for <= 5 categories
- No x/y axes, only proportions


```python
sizes = [10, 20, 30, 40]
labels = ["C", "C++", "Java", "Python"]
explode = [0.1, 0, 0, 0]
colors = ['r','g','y','b']

plt.figure(figsize=(6,6))
plt.pie(
    sizes,
    labels=labels,
    autopct='%0.1f%%',
    colors=colors,
    explode=explode,
    startangle=90,
    radius=1.2
)
plt.title('Pie Chart: Language Share')
plt.legend(loc='upper left')
plt.show()
```

#### DONUT CHART STYLE


```python
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, radius=1.2)
plt.pie([1], colors='white')  # center hole
plt.title('Donut Chart')
plt.show()
```

# 5. STEM PLOT (Discrete Signals / Sequences)

## Use Case:
- Signal processing
- Discrete time series
- x: Discrete numerical
- y: Numerical



```python
x_vals = [1,2,3,4,5]
y_vals = [2,4,6,8,4]

plt.figure(figsize=(6,4))
plt.stem(x_vals, y_vals, linefmt=':', markerfmt='ro', basefmt='g', label='Signal')
plt.xlabel('Index (Discrete)')
plt.ylabel('Value (Numerical)')
plt.title('Stem Plot')
plt.legend()
plt.show()
```

# 6. BOX & WHISKER PLOT (Distribution + Outliers)
## Use Case:
- Compare distributions
- Detect outliers
- y-axis: Numerical


```python

python_scores = [10,20,30,40,50,60,70,150]
cplusplus_scores = [10,20,30,40,50,60,70,90]

plt.figure(figsize=(6,4))
plt.boxplot(
    [python_scores, cplusplus_scores],
    widths=0.4,
    showmeans=True,
    labels=["Python", "C++"],
    notch=False
)
plt.ylabel('Scores (Numerical)')
plt.title('Box Plot Comparison')
plt.show()
```

# 7. STACK PLOT (Cumulative Trends over Time)
## Use Case:
- Show part-to-whole over time
- x: Numerical (time)
- y: Numerical areas


```python
x = [1,2,3,4,5]
area1 = [1,3,2,6,4]
area2 = [5,7,2,4,2]
area3 = [5,7,2,4,7]
labels = ['Area 1','Area 2','Area 3']

plt.figure(figsize=(6,4))
plt.stackplot(x, area1, area2, area3, labels=labels)
plt.xlabel('Time (Numerical)')
plt.ylabel('Value (Numerical)')
plt.title('Stack Plot')
plt.legend(loc='upper left')
plt.show()
```

# 8. STEP PLOT (Piecewise Constant Changes)
## Use Case:
- Algorithms
- Digital signals
- Threshold changes


```python


x = [1,2,3,4]
y = [3,1,4,3]

plt.figure(figsize=(6,4))
plt.step(x, y, where='mid', marker='o', label='Python')
plt.grid(True)
plt.xlabel('Step Index (Numerical)')
plt.ylabel('Value (Numerical)')
plt.title('Step Plot')
plt.legend()
plt.show()
```

# 9. FILL BETWEEN (Area under Curve)
## Use Case:
- Highlight region
- Confidence intervals
- Area visualization


```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.figure(figsize=(6,4))
plt.plot(x, y, color='r', label='Line')
plt.fill_between(x, y, alpha=0.3)
plt.xlabel('X (Numerical)')
plt.ylabel('Y (Numerical)')
plt.title('Fill Between Plot')
plt.legend()
plt.show()
```

# 10. SUBPLOTS (Multiple Plots in One Figure)
## Use Case:
- Dashboard style visualization
- Compare multiple plots


```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
plt.plot(x, y, color='r')
plt.title('Line Plot')

plt.subplot(2,2,2)
plt.pie([1], colors='r')
plt.title('Simple Pie')

plt.subplot(2,2,3)
plt.pie(x)
plt.title('Pie from Data')

x_cat = ['a','s','d','f']
y_cat = [10,20,30,40]
plt.subplot(2,2,4)
plt.bar(x_cat, y_cat)
plt.title('Bar Plot')

plt.tight_layout()
plt.show()
```

# 11. AXIS CONTROL & TICKS
## Use Case:
- Custom labels
- Control visible range
- Improve readability


```python

x = [1,2,3,4]
y = [4,3,2,1]

plt.figure(figsize=(6,4))
plt.plot(x, y, marker='o')
plt.xticks(x, labels=['Python', 'C', 'C++', 'Java'])  # categorical labels on x
plt.yticks([1,2,3,4])
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.xlabel('Language (Categorical)')
plt.ylabel('Rank (Numerical)')
plt.title('Axis Customization Example')
plt.show()
```
