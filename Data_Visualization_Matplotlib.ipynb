
## 1. Line Plot(Trend)

* A **Line Plot** is used to visualize the relationship between two numerical variables.
* It is mainly used to observe **trends over time or ordered data**.
* X-axis → independent variable (time, sequence, categories)
* Y-axis → dependent numerical variable (values to plot)
* Key features: can customize **line color, style, markers**, and plot **multiple lines** on the same graph.



```python
x=[1,2,3,4,5]
y=[5,10,15,18,20]
plt.plot(x,y)
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Line Plot Example")
plt.show()
```

Multiple Lines

```python
y2 = [12, 14, 22, 20, 28]
y3 = [14, 12, 12, 10, 18]

plt.plot(x, y, label="Line 1")
plt.plot(x, y2, label="Line 2")
plt.plot(x, y3, label="Line 3")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Lines Plot")
plt.legend()
plt.show()
```

## 2. Scatter Plot

A **Scatter Plot** in Matplotlib is used to visualize the relationship between **two numerical variables**.

* Each point represents one observation.
* X-axis → independent variable, Y-axis → dependent variable.
* Useful to identify **patterns, correlations, clusters, or outliers**.
* Can plot multiple groups by using **color, marker, or size**.


```python
plt.figure(figsize=(10,3))
plt.scatter(tips['total_bill'], tips['tip'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title("Bill vs Tip")
plt.show()
```

## 3. Histogram

A **Histogram** in Matplotlib is used to visualize the **distribution of numerical data**.

* X-axis → value ranges (bins), Y-axis → frequency (count of values in each bin)
* Helps understand **spread, skewness, and shape** of the data
* `bins` parameter controls the number of intervals
* Can be combined with **density=True** to normalize the histogram.


```python
plt.hist(tips['total_bill'], bins=20)
plt.title("Total Bill Distribution")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

## 4. Bar Plot
A **Bar Plot** in Matplotlib is used to compare **numerical values across categories**.

* X-axis → categorical variable, Y-axis → numerical values
* The height of each bar represents the value for that category
* Useful for **comparing averages, totals, or counts** between groups
* Can be **vertical (`plt.bar`) or horizontal (`plt.barh`)** and customized with color, labels, and width


```python
avg_tip=tips.groupby('day')['tip'].mean()
plt.bar(avg_tip.index, avg_tip.values)
plt.xlabel("Day")
plt.ylabel("Average Tip")
plt.title("Average Tip per Day")
plt.show()
```

4.1 Horizontal Bar Plot

```python
plt.barh(avg_tip.index, avg_tip.values)
plt.xlabel("Average Tip")
plt.ylabel("Day")
plt.title("Average Tip per Day")
plt.show()
```

## 5. Box Plot

A **Box Plot** in Matplotlib is used to **visualize the distribution of numerical data**.

* Shows **median, quartiles, and spread** of the data
* **Whiskers** represent variability outside the upper and lower quartiles
* **Outliers** are shown as individual points outside the whiskers
* Useful for **detecting skewness, spread, and outliers** in **categorical vs numerical** data


```python
plt.boxplot(tips["total_bill"])
plt.title("Boxplot of Total Bill")
plt.show()
```

## 6. Heatmap
A **Heatmap** in Seaborn is used to visualize data in a **matrix form using colors**.

* Color intensity represents the magnitude of values (higher = darker/stronger color)
* Commonly used to show **correlation matrices** or any 2D tabular data
* Useful for **identifying patterns, relationships, and strong correlations** between variables


```python
corr = tips.corr(numeric_only=True)

plt.imshow(corr)

plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()
```

## 8. Density / KDE Plot
A **Density Plot** or **Kernel Density Estimate (KDE)** shows the **smoothed distribution of numerical data**.

* X-axis → values of the numerical variable
* Y-axis → estimated probability density (how data is distributed)
* Unlike a histogram, it is **continuous and smooth**, making it easier to see **peaks, skewness, and spread**
* Useful for **understanding the shape of the data** and comparing distributions across groups



```python
tips['total_bill'].plot(kind='density')
plt.title("Density Plot")
plt.show()
```

## 9. Subplots

**Subplots** in Matplotlib allow you to **draw multiple plots in a single figure**.

* Useful for **comparing different plots side by side or in a grid**
* You can organize plots by **rows and columns** using `plt.subplot()`
* Each subplot can have its **own data, labels, and title**


```python

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 25, 30]
y2 = [30, 25, 20, 15, 10]

# 1 row, 2 columns, first plot
plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title("Plot 1")

# 1 row, 2 columns, second plot
plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title("Plot 2")

plt.show()
```



* `plt.subplot(rows, cols, position)` → position numbers plots **left to right, top to bottom**
* Each subplot is **independent**, so you can customize labels, color, and style individually


```python
# example 2
fig, ax = plt.subplots(1,2)

ax[0].hist(tips['total_bill'])
ax[0].set_title("Histogram")

ax[1].scatter(tips['total_bill'], tips['tip'])
ax[1].set_title("Scatter Plot")
```

## 10. Confusion Matrix
A **Confusion Matrix** is a **table used to evaluate the performance of a classification model**.

* Rows → **Actual classes**, Columns → **Predicted classes**
* Shows **how many predictions were correct or incorrect** for each class
* Key terms:

  * **True Positive (TP)** → correctly predicted positive
  * **True Negative (TN)** → correctly predicted negative
  * **False Positive (FP)** → incorrectly predicted positive
  * **False Negative (FN)** → incorrectly predicted negative
* Useful for **calculating metrics** like **accuracy, precision, recall, and F1-score**

Short: **Confusion Matrix = actual vs predicted counts for classification problems.**


```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = iris.drop("species", axis=1)
y = iris["species"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
plt.show()
```

## 11. ROC Curve

A **ROC (Receiver Operating Characteristic) Curve** is used to **evaluate the performance of a binary classification model**.

* X-axis → **False Positive Rate (FPR)**
* Y-axis → **True Positive Rate (TPR / Recall)**
* Shows the **trade-off between sensitivity and specificity** at different threshold values
* The closer the curve is to the **top-left corner**, the better the model
* **Area Under the Curve (AUC)** quantifies overall performance (1 = perfect, 0.5 = random guess)

Short: **ROC Curve = visual tool to see how well a classifier separates classes across thresholds.**


```python
from sklearn.metrics import RocCurveDisplay
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = iris.drop("species", axis=1)
y = iris["species"]

# convert to binary
y = (y == "setosa").astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

RocCurveDisplay.from_estimator(model, X_test, y_test)
plt.show()
```

## 12. Precision–Recall Curve

A **Precision–Recall (PR) Curve** is used to **evaluate a binary classifier, especially on imbalanced datasets**.

* X-axis → **Recall (True Positive Rate)**
* Y-axis → **Precision (Positive Predictive Value)**
* Shows how **precision and recall trade off** at different classification thresholds
* Useful when **positive class is rare** and accuracy alone is misleading
* The closer the curve is to the **top-right corner**, the better the model

Short: **PR Curve = visual tool to understand classifier performance focusing on positive class predictions.**


```python
from sklearn.metrics import PrecisionRecallDisplay

PrecisionRecallDisplay.from_estimator(model, X_test, y_test)
plt.show()
```
